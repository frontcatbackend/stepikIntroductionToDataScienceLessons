```python
# найти среднюю концентрацию каждого из веществ в каждом из родов (колонка genus)! 
# Для этого проведите группировку датафрэйма, сохранённого в переменной concentrations, и примените метод, 
# сохранив результат в переменной mean_concentrations.

algae = pd.read_csv("C:\\Users\\ump\\Dropbox\\ПК\\Documents\\DataScience\\algae.csv")
algae
concentrations = algae.groupby(['genus','species', 'group']).aggregate({'sucrose':'mean', 'alanin':'mean', 'citrate':'mean', 'glucose':'mean', 'oleic_acid':'mean' })
concentrations
```

```python
# чему равны минимальная, средняя и максимальная концентрации аланина (alanin) среди видов рода Fucus.
# Округлите до 2-ого знака, десятичным разделителем является точка.
import numpy as np 
one = round(algae.loc[algae.genus == 'Fucus'].aggregate({'alanin':['min', 'mean', 'max']}), 2)
two = np.around(algae.query("genus == 'Fucus'").alanin.describe().loc[['min', 'mean', 'max']].values, decimals=2)
three = algae.groupby('genus').aggregate({'alanin':['min', 'mean', 'max']})

three
```
